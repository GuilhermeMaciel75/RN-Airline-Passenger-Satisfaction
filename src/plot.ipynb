```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
bases = pd.read_csv("../results/KS_base.csv")
bases.drop(['Unnamed: 0'], axis=1, inplace=True)
```

```python
mlp = pd.read_csv("../results/KS_Mlp.csv")
# mlp.drop(['Unnamed: 0'], axis=1, inplace=True)
```

```python
rf = pd.read_csv("../results/KS_RF.csv")
# rf.drop(['Unnamed: 0'], axis=1, inplace=True)
```

```python
xgb = pd.read_csv("../results/KS_XGB.csv")
# xgb.drop(['Unnamed: 0'], axis=1, inplace=True)
```

```python
kan = pd.read_csv("../results/KS_KAN.csv")
# kan.drop(['Unnamed: 0'], axis=1, inplace=True)
```

```python
plt.plot(mlp['Unnamed: 0'], mlp['ks'], label='MLP', marker='o')
plt.plot(rf['Unnamed: 0'], rf['ks'], label='Random Forest', marker='x')
plt.plot(xgb['Unnamed: 0'], xgb['ks'], label='XGBoost', marker='s')
plt.plot(kan['Unnamed: 0'], kan['ks'], label='KAN', marker='P')

plt.title('Gráfico KS x Versão')
plt.xlabel('Versão do modelo criada pelo Optune')
plt.ylabel('Valor KS')

plt.legend()

plt.show()
```

```python
plt.boxplot([mlp['ks'], rf['ks'], xgb['ks'], kan['ks']], 
            patch_artist=True,
            tick_labels=['MLP', 'Random Forest', 'XGBoost', 'kan']
    )

plt.title('Boxplot KS x Modelos')
plt.xlabel('Modelos')
plt.ylabel('KS')

plt.show()
```

```python
plt.bar(bases['model'], bases['ks'], width=0.8, edgecolor="white", color=['#FFB6C1', '#ADD8E6', '#90EE90', '#FFDAB9'])

plt.xlabel('Modelos')
plt.ylabel('KS')

plt.ylim([0.8, 1])

plt.title('Gráfico KS x Melhores Modelos')
```
