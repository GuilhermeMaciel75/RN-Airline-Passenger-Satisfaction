```python
import pandas as pd
```

```python
SEED = 51
```

Carregando o dataset

```python
dt_train = pd.read_csv('../data/raw_data/test.csv')
dt_test = pd.read_csv('../data/raw_data/train.csv')
```

Unindo os datasets

```python
df = pd.concat([dt_train, dt_test], ignore_index=False)
```

```python
df.head()
```

Agora iremos dividir o dataset entre as classes satisfied e neutral or dissatisfied

```python
df_satisfied = df[df['satisfaction'] == 'satisfied']
df_neutral = df[df['satisfaction'] == 'neutral or dissatisfied']
```

```python
df_satisfied.drop(['Unnamed: 0'], axis=1, inplace=True)
df_neutral.drop(['Unnamed: 0'], axis=1,inplace=True)
```

Embaralhando o dataset, afim de nÃ£o haver viÃ©s de dados

```python
df_satisfied = df_satisfied.sample(frac=1, random_state=SEED, replace=False).reset_index(drop=True)
df_neutral = df_neutral.sample(frac=1, random_state=SEED, replace=False).reset_index(drop=True)
```

Dividindo os dados em test, validaÃ§Ã£o e treino. Iremos pegar as 50% primeiras linhas para treinamento, depois as prÃ³ximas 25% para validaÃ§Ã£o e os Ãºltimos 25% para teste

```python
def split_data_set(df : pd.DataFrame) -> list[pd.DataFrame, pd.DataFrame, pd.DataFrame]:
    qtd_lines = df.shape[0]

    train = df.iloc[:int(qtd_lines * 0.5)]
    validation = df.iloc[int(qtd_lines * 0.5) : int(qtd_lines * 0.75)]
    test = df.iloc[int(qtd_lines * 0.75):]

    print(f"Size train: {len(train)}\nSize validation: {len(validation)}\nSize test: {len(test)}")

    return train, validation, test
```

```python
df_satisfied_train, df_satisfied_validation, df_satisfied_test = split_data_set(df_satisfied)
```

```python
df_neutral_train, df_neutral_validation, df_neutral_test = split_data_set(df_neutral)
```

Agora, temos que equilibrar os datasets de treino e validaÃ§Ã£o, assim iremos realizar um Random oversampling no dataset de treinamento e validaÃ§Ã£o da classe 'satisfied'

```python
def random_oversampling(num_samples : int, df : pd.DataFrame) -> pd.DataFrame:
    num_new_samples = num_samples - df.shape[0]

    df_samples = df.sample(n=num_new_samples, replace=True, random_state=SEED)
    df = pd.concat([df, df_samples], ignore_index=False)
    df = df.sample(frac=1, random_state=SEED, replace=False).reset_index(drop=True)

    return df
```

```python
df_satisfied_train = random_oversampling(df_neutral_train.shape[0],df_satisfied_train)
print(f"New train set size: {df_satisfied_train.shape[0]}")
```

```python
df_satisfied_validation = random_oversampling(df_neutral_validation.shape[0],df_satisfied_validation)
print(f"New validation set size: {df_satisfied_validation.shape[0]}")
```

Unindos os csv's para ter o conjunto final de treinamento, test e validaÃ§Ã£o

```python
# Save test
df_train = pd.concat([df_satisfied_train, df_neutral_train], ignore_index=False)
df_train = df_train.sample(frac=1, random_state=SEED, replace=False).reset_index(drop=True)
df_train.to_csv('../data/clean_data/train.csv', index=False)

# Save validation
df_validation = pd.concat([df_satisfied_validation, df_neutral_validation], ignore_index=False)
df_validation = df_validation.sample(frac=1, random_state=SEED, replace=False).reset_index(drop=True)
df_validation.to_csv('../data/clean_data/validation.csv', index=False)

# Save test
df_test = pd.concat([df_satisfied_test, df_neutral_test], ignore_index=False)
df_test = df_test.sample(frac=1, random_state=SEED, replace=False).reset_index(drop=True)
df_test.to_csv('../data/clean_data/test.csv', index=False)
```
